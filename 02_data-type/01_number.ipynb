### 숫자형의 연산

```python
num1 = 12
num2 = 31

print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2)
```

```python
print(num1 ** num2) # 제곱 -> 12의 31제곱
print(num1 // num2) # 나누기 후의 몫
print(num1 % num2) # 나누기 후의 나머지
```
