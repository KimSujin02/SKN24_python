```python
str1 = "오늘의 점심 메뉴는"
str2 = "김치볶음밥"

print(str1 + str2) # 문자열에서의 사칙연산은 불가능. +는 문자열을 이어주는 역할을 함
# print(str1 - str2) # 타입에러
# print(str1 * str2) # 타입에러
# print(str1 / str2) # 타입에러
```

### f-string 포맷팅

```python
# 다람쥐는 겨울철 양식으로 도토리를 100개 컬리했습니다.
user = '다람쥐'
food = '도토리'
count = 100

# f-string 사용법
print("f-string으로 출력하기")
print(f"{user}는 겨울철 양식으로 {food}를 {count}개 컬리했습니다.")


print("format 함수로 출력하기")
print("{0}는 겨울철 양식으로 {1}를 {2}개 컬리했습니다.".format(user, food, count))
print("{x}는 겨울철 양식으로 {y}를 {z}개 컬리했습니다.".format(x=user, y=food, z=count))

print("콤마로 출력하기")
# 이렇게 해도 됨. 근데 콤마로 하면 띄어쓰기가 나옴!! 그래서 이런식으로 출력할때는 sep이 필요함.
print(user, "는 겨울철 양식으로 ", food, "를 ", count, "개 컬리했습니다.")
# sep의 풀단어는 : separator 구분자!라는 뜻
print(user, "는 겨울철 양식으로 ", food, "를 ", count, "개 컬리했습니다.", sep='!!!!!')

print("+로 출력하기")
print(user+"는 겨울철 양식으로 "+food+"를 "+str(count)+"개 컬리했습니다.")
```
