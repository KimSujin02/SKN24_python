### 논리형의 연산

```python
print('hello' == 'Hello')   # False
print(100 == 100)           # True
print('hello' != 'Hello')   # True
print(10 > 9)               # True
print(10 < 9)               # False
print(10 >= 9)              # True
print(10 <= 9)              # False

if(bool1) :
    print('bool1은 True')

if(bool2) :
    print('나오면 안됨')
else :
    print('bool2은 False')
    
print('a' < 'A ')
# 아스키 코드표를 보면 대문자는 소문자보다 먼저임. = 수의 크기가 작음
```

```python
print((1 == 1) and (1 == 2))    # F
print((1 == 1) or (1 == 2))     # T
print(not (1 == 2))             # T
```
