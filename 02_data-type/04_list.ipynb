# 리스트(배열)

```python
avatar = ['제이크', '네이터리', '그레이스', '마일즈']
nothing = []
anything = [2026, 1, 2, 'python', ['number', 'string', True], False]
```

```python
type(anything)
```

```python
# 형변환 함수 list()
rainbow_str = '빨 주 노 초 파 남 보'
rainbow_list = list(rainbow_str)
print(rainbow_list)     # ['빨', ' ', '주', ' ', '노', ' ', '초', ' ', '파', ' ', '남', ' ', '보']

rainbow_list = rainbow_str.split('-')
print(rainbow_list)     # ['빨', '주', '노', '초', '파', '남', '보']
print(len(rainbow_list))
```

### 리스트의 연산

```python
added_list = avatar + anything
```

```python
multiple_list = avatar * 10
```

```python
print(added_list)
print(multiple_list)
```

### 리스트의 메서드

```python
num_list = [2026, 1, 2, 11, 3]
print(num_list)

num_list.append('오전')
print(num_list)

num_list.insert(0, 777)
print(num_list)
```

```python
num_list.remove('오전')      # 해당 요소를 삭제함. 값.
print(num_list)
```

```python
num_arr = [1,2,3,4,5,6,1]
num_arr.remove(1)       # 중복되는 요소가 있는 경우, 그냥 앞에서 하나 찾는거 같음
print(num_arr)
```

```python
# num_list.append('오전')

# 만약 자료형이 다른 값들이 섞여있으면 Error 발생함.
# TypeError 발생함.
num_list.sort()
print(num_list)
```

```python
num_list.sort(reverse=True)
print(num_list)
```

```python
avatar = ['제이크', '네이터리', '그레이스', '마일즈']
print(sorted(avatar))
print(avatar)
# sort는 리스트 자체의 값을 바꿈
# sorted는 걍 결과물만 반환하고 원래 변수의 값은 바뀌지 않음.
```

### ⭐index (인덱싱과 슬라이싱)

```python
# anything = [2026, 1, 2, 'python', ['number', 'string', True], False]

first_else = anything[1]
forth_else = anything[4]
print(first_else)       # 인덱스 1번의 내용을 출력
print(forth_else)       # 인덱스 4번의 내용을 출력

print(anything[-1])     # 인덱스 맨 뒤의 내용을 출력
```

```python
anything[-1] = True
print(anything)
```

```python
print(anything[1:3])
```
