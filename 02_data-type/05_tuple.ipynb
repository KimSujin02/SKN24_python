# 튜플

```python
empt_tuple = ()     # tuple이긴한데 의미가 없음
empt_tuple = (1)     # int로 나옴
empt_tuple = (1,)     # tuple로 인식
num_tuple = (2026, 1, 2, 14, 1)
anything = (1, 'python', ['tuple', 'is', '불변'], True)
```

```python
type(empt_tuple)
```

```python
# ★★★★★ 튜플의 특징 = 불변 ★★★★★

print(num_tuple[2])

# anything[-1] = False # TypeError 발생함. 튜플은 불변하기 때문~!!

tuple_list = anything[2]        # ['tuple', 'is', '불변']
tuple_list[2] = 'immutable'     # ['tuple', 'is', 'immutable']
print(tuple_list)

print(anything)                 # anything[2]의 내용이 바뀌어 있음.
```
