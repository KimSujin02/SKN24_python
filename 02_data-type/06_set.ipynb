### 셋의 집합 연산

```python
fruits = {'apple', 'banana', 'orange', 'watermelon'}
lunch = {'apple', 'orange', 'carrot', 'chiken'}
```

```python
# 합집합 union

print(fruits.union(lunch)) # 중복이 제거되어 합쳐짐!
print(fruits | lunch)
```

```python
# 교집합 intersection

print(fruits.intersection(lunch))       # 둘이 겹쳐지는 값만 나옴!
print(fruits & lunch)
```

```python
# 차집합 difference

print(fruits.difference(lunch))     # fruits에서 lunch값 빼기
print(lunch.difference(fruits))     # lunch에서 fruits값 빼기
print(fruits - lunch)
```

```python
# 대칭차집합 symmetric_difference

print(fruits.symmetric_difference(lunch))
print(lunch.symmetric_difference(fruits))
print(fruits ^ lunch)
```
