# 딕셔너리

```python
my_dict = {
    'name' : 'kimsujin',
    'height' : 156,
    'job' : 'developer',
    'bag' : ['note', 'pouch', 'water'],
    'happy' : True
    }
```

```python
print(my_dict.keys())
print(my_dict.values())

print(my_dict.items())
```

```python
# dictionary 형식에서 값 가져오는 방식 1
print(my_dict['name'])

my_dict['name'] = '김아무개'

# dictionary 형식에서 값 가져오는 방식 2
print(my_dict.get("name"))
print(my_dict.get("bag")[1])
```

```python
my_dict['hobby'] = '게임'
print(my_dict)
```

```python
del my_dict['bag']
print(my_dict)
```

```python
# 1. swallow copy 얕은 복사
a = {'a' : 1, 'b' :2, 'c' : 3}
b = a # a의 주소값을 공유함.
print(a)
print(b)

a['a'] = 100
print(a)
print(b) # 같이 바뀜
```

```python
# 2. deep copy
c = {'a' : 1, 'b' :2, 'c' : 3}
d = c.copy()    # c의 값을 물려받은? 새로운 주소값 생성해서 d에 대입함
print(c)
print(d)

c['a'] = 100
print(c)    # 얘 혼자 바뀜
print(d)       # 바뀌지 않음
```
