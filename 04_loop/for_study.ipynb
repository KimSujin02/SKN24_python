```python
# 구구단을 출력해보자
for n in range(2, 10) :     # 2에서 9단까지
    print(n, "단")
    for m in range(1, 10) :     # 1에서 9곱하기 까지
        print(n, "X", m, "=", n*m)
    print()

# 끝~~~
```

```python
# 별찍기를 해보자 1
# *****
# ****
# ...

a = 5
for n in range(a) :
    print("*" * (a-n))
```

```python
#      *
#     **
#    ***
#   ****
#  *****
a = 6
for i in range(1, a):
    print(" " * (a - i), end="")
    print('*' * i)
```

```python
# 별찍기를 해보자 2
#    *
#   ***
#  *****
#   ***
#    *

a = 7
for i in range(a//2):       # 몫만큼 돌기
    print(' ' * (a//2 - i), end = '')
    print('*' * (2*i+1))

for i in range(a//2-1):       # 몫-1만큼 돌기
    print(' ' * (i + 2), end = '')
    print('*' * ((a//2*2)-3-2*i))
```
