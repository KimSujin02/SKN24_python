## **문제 1: 손님 응대하기**

**"다람쥐는 카페에서 아르바이트를 시작했다."**

## **문제 8: 손님이 그만 올 때까지 기다리기**

**"손님이 그만 올 때까지 카운터에서 기다리는 다람쥐."**

다람쥐는 손님이 그만 올 때까지 카운터를 지켜야 합니다.

1. 변수 `customers`에 초기값 `5`를 설정하세요.
2. `while`문을 사용해 손님이 줄어들 때마다 `customers`를 1씩 감소시키세요.
3. 손님이 0이 될 때까지 **"남은 손님 수: [customers]명"**을 출력하세요.
4. 손님이 모두 떠나면 **"오늘의 영업이 끝났습니다!"**를 출력하세요.


```python
customers = 5
while customers >= 0 :
    print(f"남은 손님 수 : {customers}명")
    if customers == 0 :
        print("오늘의 영업이 끝났습니다!")
    customers -= 1
```

## **문제 9: 알바비 계산하기**

**"다람쥐는 일한 시간만큼 알바비를 계산합니다."**

다람쥐의 시급은 **1시간에 1000원**입니다.

1. 변수 `work_hours`에 일한 시간(정수)을 입력받으세요.
2. `while`문을 사용해 일한 시간만큼 반복하며 `pay`에 시급을 더하세요.
3. 최종 알바비를 **"오늘의 알바비는 [pay]원입니다."**로 출력하세요.

```python
work_hours = int(input("일한 시간(정수)를 입력해주세요 : "))

pay = 0

while work_hours > 0 :
    pay += 1000
    work_hours -= 1
print(f"오늘의 알바비는 {pay}원 입니다.")
```

## **문제 10: 손님 응대 중단하기**

**"다람쥐는 갑자기 피곤해져서 손님 응대를 중단하고 싶다."**

다람쥐는 카페에서 손님을 맞이하고 있습니다. 하지만 다람쥐는 너무 피곤해서 **3번째 손님**까지만 응대하고 싶습니다.

1. 손님 리스트 `customers`에 `'토끼', '곰', '여우', '호랑이', '늑대'`를 저장하세요.
2. `for`문을 사용해 각 손님을 응대하며 다음 메시지를 출력하세요:
    - **"안녕하세요, [손님 이름]님! 무엇을 도와드릴까요?"**
3. 3번째 손님까지 응대한 후, 반복문을 종료하고 **"죄송합니다, 오늘은 여기까지입니다!"**를 출력하세요.


```python
customers = ['토끼', '곰', '여우', '호랑이', '늑대']

count = 0
for n in customers :
    if count == 3 :
        print("죄송합니다, 오늘은 여기까지 입니다.")
        break
    print(f"안녕하세요, {n}님! 무엇을 도와드릴까요?")
    count += 1

print("===================다른 방법 enumerate===================")
# enumerate는 시퀀스의 idx와 value를 list로 반환해줌.
for n, customer in enumerate(customers) :
    if n == 3 :
        print("죄송합니다, 오늘은 여기까지 입니다.")
        break
    print(f"안녕하세요, {customer}님! 무엇을 도와드릴까요?")
```

## **문제 11: 주문 실수 중단하기**

**"다람쥐는 주문을 받다가 실수를 발견하면 즉시 중단하려고 한다."**

다람쥐는 손님들의 주문을 확인하고 있습니다. 주문에 문제가 있으면 즉시 중단해야 합니다.

1. 주문 리스트 `orders`에 `'아메리카노', '카페라떼', '핫초코', '모르겠어요', '에스프레소'`를 저장하세요.
2. `for`문을 사용해 각 주문을 확인하며 다음과 같이 출력하세요.
    - **"[음료 이름] 주문 확인했습니다!"**
3. 만약 주문이 `'모르겠어요'`이면, **"주문에 문제가 있습니다! 확인이 필요합니다."**를 출력하고 반복문을 종료하세요.


```python
orders = ['아메리카노', '카페라떼', '핫초코', '모르겠어요', '에스프레소']
for n in orders :
    if n == '모르겠어요' :
        print('주문에 문제가 있습니다! 확인이 필요합니다.')
        break
    print(f'{n} 주문 확인했습니다!')
```
