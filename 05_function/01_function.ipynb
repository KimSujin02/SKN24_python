# 함수

```python
# 이것은 파라미터도 없고 return 값도 없는 함수임.
def print_dinner() :
    print("오늘 저녁은 치킨입니다!")
    return
```

```python
print_dinner()

result = print_dinner()
print(result) # 반환되기는 해도 반환되는 값이 없어서 None이라고 나옴
```

```python
# 파라미터 (X) return값(O) 함수
# 오늘 저녁은 피자 입니다 문자열을 반환받아 출력
def return_dinner() :
    return '오늘 저녁은 피자입니다.'

print(return_dinner())
```

```python
# 파라미터 (O) return값 (X) 함수
# 오늘 저녁 메뉴를 menu 파라미터로 전달하고 '오늘 저녁은 {menu}입니다' 출력
def print_my_dinner(menu) :
    print(f'오늘 저녁은 {menu}입니다.')

print_my_dinner('삼겹살')
```

```python
# 파라미터 (O) return (O) 함수
# 오늘 저녁 메뉴를 menu 파라미터로 전달하고 '오늘 저녁은 {menu}입니다' 문자열을 반환받아 출력
def return_my_dinner(menu) :
    return f'오늘 저녁은 {menu}입니다.'

print(return_my_dinner('스테이크'))
```

### 위치 인자와 키워드 인자

```python
user_list = [{"hobby":"게임하기", "name":"김수진", 'emotion':'funny'}
             , {"hobby":"잠자기", "name":"김김김", 'emotion':'soso'}
             ,{"hobby":"유튜브보기", "name":"수수수", 'emotion':'tired'}]

def return_hobby(name, hobby, emotion='happy') :
    return f'{name}님의 은밀한 취미생활은 {hobby}입니다. 기분은 {emotion}'
```

```python
# 위치 인자 연습
for n in user_list :
    print(return_hobby(n['name'], n['hobby'], n['emotion']))
```

```python
# 키워드 인자 연습1
for n in user_list :
    print(return_hobby(hobby = n['hobby'], name = n['name'], emotion = n['emotion']))
```

```python
# 키워드 인자 연습2
# 키워드 인자 한 번 쓰면 그 뒤에 있는 인자도 무조건 키워드 인자로 진행해야됨!
for n in user_list :
    print(return_hobby(n['name'], hobby = n['hobby'], emotion = n['emotion']))
```

### [참고] Call By Value & Call By Reference

```python
def change_num(num):
    num = 999
    print("함수 안에서 num:", num)
```

```python
arg_num = 15
print(arg_num)
change_num(arg_num)
print(arg_num)
```

```python
def change_arr(arr):
    # arr = arr.copy()
    # 리스트나 딕셔너리 자료형은 원본 데이터에 영향을 주지 않도록 deep copy를 사용해야됨.
    arr[0] = "Call By Reference는 이래요."
    print('함수 안에서 arr : ', arr)
```

```python
arg_arr = [2026, 1, 5]
print(arg_arr)
change_arr(arg_arr)
print(arg_arr)
```
