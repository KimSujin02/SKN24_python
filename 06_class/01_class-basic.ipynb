# 클래스

```python
class Car:
    country = 'korea'
    def __init__(self, name, price) :
        self.name = name
        self.price = price
        
    def ride(self) :
        print("말~달리자~~..... : ", self.name)
        print(f"{self.price}만원 짜리")
        print(f"{self.country}에서 달려요")
```

```python
a = Car("막시무스 3세", "100")
a.ride()
```

```python
class Student :
    # 클래스 속성
    dream = '취업'
    # 생성자 => 없어도 됨. 클래스에 인자값을 주고 싶지 않다면 안 줘도 됨!
    def __init__(self, name, age, email) :
        self.name = name
        self.age = age
        self.email = email
        print('생성자 호출됨.')
    # 메서드
    def study(self) :
        print('오늘도 열심히 공부하는 우리 24기!')
    
    def student_info(self) :
        print(f'{self.name}님 하이요. 나이는 {self.age}입니다.')
        print(f'메일 주소는 {self.email}이네요.')
```

```python
# 클래스명() => 생성자를 호출 => 인스턴스를 생성하고 주소값을 반환
# 그럼 즉 student라는 변수에는 해당 인스턴스의 주소값을 가지고 있는 변수인것임!
student = Student('김수진', '25', 'selina9135@gmail.com')

# 생성자 호출 시마다 메모리에 새로운 인스턴스를 할당하므로, 인스턴스의 주소값이 달라짐.
print(student) # 이것이 인스턴스 생성 된 거 주소값 출력한거임.

print(student.dream)
print(student.study)
print(student.study()) # None이 나오는 이유 : return값이 없어서!
```

```python
student.student_info()

student.dream = '머기업 상무님'

print(f'제 목표는 {student.dream}입니다.')
```

### 클래스 속성 vs 인스턴스 속성

```python
class Student:
    dream = ['취업']
    name = '아무개'
    
    def study(self) :
        print('오늘도 열심히 공부하는 24기!')
    
    def introduce(self) :
        print(f'저는 {self.name}이구요, 제 꿈은 {self.dream}입니다!')
```

```python
student1 = Student()

# student1 인스턴스의 name 공간은 인스턴스 속성임.
student1.name = '김수진'

# student1 인스턴스의 dream 공간은 클래스 속성이라서 dream의 arr 공간을 찾아서 속성에 값을 대입해주는거임
student1.dream.append('로또당첨돼서 취미로 회사다니는 대기업 상무님')
student1.dream.append('억만장자')
student1.introduce()

########################

student2 = Student()

student2.name = '김아무개'
student2.introduce()
```
