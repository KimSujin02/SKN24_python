# 상속

```python
class Person:
    national = 'korea'
    
    def greeting(self) :
        return 'Hello, This is Python!'
class Student(Person):
    def greeting(self):
        return '안녕하세요, 저는 학생입니다.'
    def change_national(self, value) :
        self.national = value
```

```python
student = Student()
print(student.national)
print(student.greeting())

person = Person()
print(person.greeting())

print(student.national)
student.change_national('america')
print(student.national)
```

```python
class Person:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
        
    # 이 함수를 선언하기 전에 기본적으로 들어있는 __str__의 기본 return값은 해당 인스턴스(생성된거)의 주소값임.
    # 그럼 이거는 class라는 기본 형식 부모의 __str__메서드를 오버라이딩 한거임!
    def __str__(self) :
        return f'<Person {self.name}, {self.phone}>'
```

```python
person = Person('김수진', '010-1111-2222')

print(person)
```

```python
class Teacher(Person) :
    # 상속 받았을때는 부모가 가진 생성자에서 들어가야되는 매개변수는 꼭 선언해줘야됨.
    # 자식 입장에서는 !
    def __init__(self, name, phone, major) :
        self.major = major
        
        # 1. 이렇게 부모의 인스턴스 속성을 직접 쓰는 방법이 있고,
        # self.name = name
        # self.phone = phone
        
        # 2. 이렇게 부모 클래스명으로 접근해서 인스턴스 속성의 값을 설정해주는 방법이 있음.
        # Person.__init__(self, name, phone)
        
        # 3. 그리고 슈퍼함수 사용해서 선언하는 방법도 있음
        # 근데 이거는 self를 넣지 않아도 됨.
        # 왜냐 : 부모 클래스를 아예 선언한거임!! Person().메서드명 이렇게 실행한거나 다름이 없음.
        super().__init__(name, phone)
```

```python
teacher = Teacher('김수진', '010-1111-2222', 'AI')
print(teacher)
print(teacher.major)
```

### 다중 상속

```python
class Person:
    fingers = 10
    eyes = 2
    
    def introduce(self) :
        return '저는 사람입니다!'
class Animal:
    fur = True
    def introduce(self) :
        return '저는 동물입니다. ㅇㅅㅇ'

class BanInBanSu(Person, Animal) :
    pass
```

```python
a = BanInBanSu()

print(a.fingers, a.eyes, a.fur)
# 여기에서 person의 introduce의 값이 나오는 이유는
# 동일한 이름의 함수를 찾으면 앞에 있는 class의 메서드를 찾는거임! 
print(a.introduce())
```

### 오버라이딩

```python
class Animal:
    species = '동물'
    def introduce(self) :
        print('동물의 왕국에 오신 것을 환영합니다.')

class Mammalia(Animal) :
    species = '포유류'
    def has_fur(self) :
        return True

class Squirrel(Mammalia) :
    name = '다람쥐'
    def eat(self):
        print('도토리는 정말 맛있다!')

class SquirrelTeacher(Squirrel):
    species = '인간'
    def has_fur(self) :
        return False
    def introduce(self):
        print('안녕하세요 다람쥐 강사입니다.')
```

```python
animal = Animal()
animal.introduce()
print(f'{animal.species}입니다.')

print()

mamal = Mammalia()
mamal.introduce()
print(f'{mamal.species}입니다.')
print('털을 가지고 있나요? : ', mamal.has_fur())

print()

squirrel = Squirrel()
squirrel.introduce()
print(f'{squirrel.species}입니다.')
print('털을 가지고 있나요? : ', squirrel.has_fur())
squirrel.eat()

print()

teacher = SquirrelTeacher()
teacher.introduce()
print(f'{teacher.species}입니다.')
print('털을 가지고 있나요? : ', teacher.has_fur())
```

```python
# mro의 뜻 : method resolution order
# 메서드를 찾아가는 순서임.
# SquirrelTeacher > Squirrel > Mammalia ... 등
# 근데 제일 밑에 잇는 object는 무엇인가..
    # 클래스의 기본 메서드를 가지고 있는 내장 오브젝트?
SquirrelTeacher.__mro__
```
