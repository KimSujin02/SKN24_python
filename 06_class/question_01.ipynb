### **문제 1: 클래스 속성, 인스턴스 속성 테스트**

1. 클래스 속성 `species`를 설정하여 "다람쥐"로 초기화하세요.
2. 생성자를 통해 인스턴스 속성 `name`과 `current_job`을 설정하세요.
3. 메서드 `introduce`를 만들어 이름과 현재 직업을 소개하는 문자열을 반환하세요.
4. 다른 인스턴스를 생성해도 `species`는 공유되지만, `name`과 `current_job`은 서로 독립적임을 확인하세요.

```python
class Test1:
    species = '다람쥐'
    
    def __init__(self, name, current_job):
        self.name = name
        self.current_job = current_job
        print(self.name)
    
    def introduce(self) :
        print(f'안녕하세요. 제 이름은 {self.name}이고, 현재 직업은 {self.current_job}입니다.')

a = Test1('김수진', '백수')
a.introduce()
print(a.species)

b = Test1('김김김', '억만장자')
b.introduce()
print(b.species)
```

### **문제 2: 클래스 속성과 인스턴스 속성의 차이**

다람쥐는 이직 준비를 하며 다양한 직업 스킬을 학습 중입니다. 클래스 속성과 인스턴스 속성의 차이를 확인하세요.

1. `Character` 클래스를 생성하세요.
2. 클래스 속성 `skills`를 정의하여 빈 리스트로 초기화하세요.
3. 생성자에서 인스턴스 속성 `skills`를 별도로 초기화하세요.
4. `add_skill` 메서드에서 클래스 속성이 아닌 인스턴스 속성에 스킬을 추가하도록 수정하세요.


```python
class Character:
    skills = ['class']
    
    def __init__(self):
        self.skills = []
    
    def add_skill(self, skill) :
        skills = []
        skills.append(skill)
        print("인스턴스 변수에 있는 skills : ", skills)
        print("클래스 변수에 있는 skills : ", self.skills)
    
    def add_skill_global(self, skill) :
        self.skills.append(skill)
        print("skills : ", self.skills)

a = Character()
a.add_skill('자바')

a.add_skill_global('자바')
a.add_skill_global('파이썬')


a.add_skill('장고')
```


### **문제 3: 인스턴스 속성과 클래스 속성의 우선순위**

다람쥐는 이직을 위해 별명을 여러 개 사용합니다. 클래스 속성으로 `nickname`을 설정하지만, 각 인스턴스마다 별명을 변경할 수 있습니다.

1. `Character` 클래스에 클래스 속성 `nickname`을 설정하세요.
2. 생성자에서 인스턴스 속성 `nickname`을 설정하세요.
3. 클래스 속성과 인스턴스 속성이 충돌할 때, 인스턴스 속성이 우선됨을 확인하세요.

```python
name = '안녕하세요'
class Character:
    nickname = '나 클래스'
    
    def __init__(self) :
        nickname = '나 인스턴스'
        print(nickname)
    
    def global_test(self, new_name) :
        global name
        print('글로벌 변수의 값을 바꾸고 싶다!!! 전: ', name)
        name = new_name
        print('후 :', name)

a = Character()
print(a.nickname)

a.global_test('김수진')
```


### **문제 4: 네임스페이스 테스트 - 전역과 로컬**

다람쥐는 이직 준비 도구를 사용하며, 전역 변수와 로컬 변수를 관리합니다. `namespace_test` 함수를 작성해 네임스페이스를 테스트하세요.

1. 전역 변수 `tool`을 "연필"로 설정하세요.
2. 함수 `namespace_test` 내부에 로컬 변수 `local_tool`을 "키보드"로 설정하고, `global` 키워드를 사용해 전역 변수를 "펜"으로 수정하세요.
3. 함수 내부에서 로컬 변수와 전역 변수의 값을 출력하세요.


```python
tool = '연필'

def namespace_test() :
    local_tool = '키보드'
    global tool
    tool = '펜'
    print('함수 내부에서의 로컬변수 :', local_tool)
    print('함수 내부에서의 전역변수 :', tool)

print("원래 전역변수 tool의 값은", tool)
namespace_test()
print("함수 실행 후의 전역변수 tool의 값은", tool)
print("값이 바뀌었음을 알 수 있음!")
```

### **문제 5: 네임스페이스 테스트 - 클래스와 인스턴스**

다람쥐는 이직 준비 중 클래스를 통해 도구를 관리합니다. `Toolbox` 클래스를 만들어 네임스페이스를 테스트하세요.

1. 클래스 속성 `tools`를 ["노트북", "책상"]으로 초기화하세요.
2. 생성자를 통해 인스턴스 속성 `tools`를 설정하고, 객체마다 다른 도구를 가질 수 있도록 설정하세요.
3. `add_tool` 메서드를 만들어 클래스 속성에 새 도구를 추가하세요.


```python
class Toolbox:
    tools = []
    def __init__(self):
        print("인스턴스 생성하고 tools의 값 :", self.tools)
    
    def add_tool(self, new_tool) :
        self.tools.append(new_tool)
    
    # 이거 하면 클래스를 공유하는 모든 인스턴스의 tools의 값이 바뀌는 거임.
    def add_tool_class(self, new_tool) :
        __class__.tools.append(new_tool)
        print("Toolbox 클래스의 tools가 아예 바뀜:", __class__.tools)

a = Toolbox()

a.add_tool_class('공책')

print()

b = Toolbox()
print("b가 생성한 toolbox가 가지고 있는 변수인 tools의 값 : ", b.tools)
print(Toolbox.tools)


a.add_tool('지우개')
a.add_tool('가위')
print('a의 tools : ', a.tools)
```

```python
# 데코레이터를 이용해 클래스 메서드로 만들고 클래스 속성에 접근
class Toolbox:
    tools = ['노트북', '책상']
    def __init__(self, tool):
        pass
    
    # 이거쓰면 self로 받던것이 class가 들어감. 클래스 속성을 변경하기 위해 사용함.
    # 그러면 이 메서드는 그냥 메서드가 아니라 클래스 메서드가 되는 거임.
    @classmethod
    def add_tool(cls, tool) :
        cls.tools.append(tool)

test = Toolbox('')
test.add_tool("웋히히ㅣ")
test.add_tool("웋히히ㅣ2")
print(test.tools)

test2 = Toolbox('')
print(test2.tools)
```

### **문제 6: 네임스페이스 테스트 - nonlocal**

다람쥐는 이직 준비 중 추가적인 정보가 필요합니다. 중첩 함수와 `nonlocal` 키워드를 사용해 바깥 함수의 변수를 수정하세요.

1. 함수 `nested_function`을 작성하세요.
2. 바깥 함수의 변수 `status`를 "작성 중"으로 초기화하세요.
3. 중첩 함수에서 `status`를 "완료"로 수정하세요.

```python
def nested_function1() :
    status = '작성중'
    def nested_function2() :
        nonlocal status
        status = '완료'
    print(status)
    nested_function2()
    print(status)

nested_function1()
```
