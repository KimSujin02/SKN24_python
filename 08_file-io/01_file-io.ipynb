# 파일 입출력(File Input/Output)

```python
# r: 읽기 모드
# w: 쓰기 모드 (덮어쓰기)
# a: 수정 모드 (파일 끝에 내용 추가하기)
 # 파이썬도 파일 열면 꼭 close()해야됨. 락걸릴 수 있음
f1 = open("test.txt", 'a', encoding='utf-8')
f1.write('오늘의 점심 메뉴 : 1\n')
f1.write('오늘의 점심 메뉴 : 안동 찜닭과 볶음 짬뽕\n')
f1.close()
```

```python
f2 = open("our_song.txt", 'w', encoding='utf-8')

our_song = [
    "공주의 규칙\n\n"
    ,'나를 사랑하기\n'
    ,'남도 사랑하기\n'
    ,'삶을 사랑하기\n'
    ,'여유 잃지않기\n'
    ,'미소 잃지 않기'
]
f2.writelines(our_song)

f2.close()
```

```python
f3 = open('our_song.txt', 'r')

# f3_read = f3.read()

# read()후에 파일의 내용 전체를 다 꺼내 왔기 때문에 또 readline()하면 에러가남!
# f3_readline = f3.readline()

f3_readlines = f3.readlines()

# print("f3.read() : ", f3_read, "/", type(f3_read))
# print("f3.readline() : ", f3_readline, "/", type(f3_readline))
print("f3.readlines() : ", f3_readlines, "/", type(f3_readlines))

f3.close()
```
