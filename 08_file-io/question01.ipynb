
### **문제 2: `.csv` 파일 생성 및 읽기**

다람쥐는 이직 준비를 위해 지원했던 회사들의 정보를 저장하려고 합니다.

1. `applications.csv` 파일을 생성하고 다음 데이터를 작성하세요.
    
    ```
    회사명,직무,결과
    회사A,데이터 분석가,합격
    ```


### **문제 3: `.log` 파일 읽기**

다람쥐는 프로그램 실행 로그를 확인하려고 합니다.

1. `system.log` 파일을 다음 내용으로 생성하세요.
    
    ```
    [INFO] 2024-01-01 12:00:00: 시스템 시작
    [WARN] 2024-01-01 12:05:00: 메모리 사용량 증가
    [ERROR] 2024-01-01 12:10:00: 프로그램 종료
    
    ```
    
2. 파일을 읽어와 `WARN` 또는 `ERROR`가 포함된 줄만 출력하세요.


````python
file = open('system.log', 'w')

list = [
    "[INFO] 2024-01-01 12:00:00: 시스템 시작\n"
    , "[WARN] 2024-01-01 12:05:00: 메모리 사용량 증가\n"
    , "[ERROR] 2024-01-01 12:10:00: 프로그램 종료"
]

file.writelines(list)
file.close()

file = open('system.log', 'r')

for n in file.readlines() :
    if ('WARN' in n) or ('ERROR' in n) :
        print(n, end='')
file.close()
````


### **문제 4: `.json` 파일 생성 및 읽기**

다람쥐는 좋아하는 음식 데이터를 저장하고 싶습니다.

1. `favorite_foods.json` 파일을 생성하고 다음 데이터를 작성하세요.
    
    ```json
    {
        "name": "다람쥐",
        "favorite_foods": ["도토리", "호두", "밤"]
    }
    
    ```
    
2. 파일을 읽어와 다람쥐의 좋아하는 음식 리스트를 출력하세요.


````python
file = open('favorite_foods.json', 'w', encoding='utf-8')

file.write("""{
        "name": "다람쥐",
        "favorite_foods": ["도토리", "호두", "밤"]
    }""")
file.close()

file = open('favorite_foods.json', 'r')

contents = eval(file.read())
print(contents['favorite_foods'])
file.close()
````

````python
# 내장모듈인 json을 활용해서 가져오는 방법
import json
file = open('favorite_foods.json', 'w', encoding='utf-8')

file.write("""{
        "name": "다람쥐",
        "favorite_foods": ["도토리", "호두", "밤"]
    }""")
file.close()

file = open('favorite_foods.json', 'r')

contents = json.load(file)
print(contents['favorite_foods'])
file.close()
````

### **문제 5: `.html` 파일 수정**

다람쥐는 자신의 이력서를 작성 중입니다. `resume.html` 파일을 다음 내용으로 생성하고 수정하세요.

1. `resume.html` 파일 생성 내용:
    
    ```html
    <html>
        <head><title>다람쥐의 이력서</title></head>
        <body>
            <h1>다람쥐의 이력서</h1>
            <p>경력: 신입</p>
        </body>
    </html>
    
    ```
    
2. 수정 작업:
    - `<p>` 태그 안의 "신입"을 "1년 경력"으로 변경하세요.
3. 수정된 파일을 읽어와 전체 내용을 출력하세요.

````python
file = open('resume.html', 'w')
file.write("""
    <html>
        <head><title>다람쥐의 이력서</title></head>
        <body>
            <h1>다람쥐의 이력서</h1>
            <p>경력: 신입</p>
        </body>
    </html>
""")
file.close()

file = open('resume.html', 'r')
contents = file.read().replace('신입', '1년 경력')
file.close()

file = open('resume.html', 'w')
file.write(contents)
file.close()

file = open('resume.html', 'r')
print(file.read())
file.close()
````
