# 예외(Exception)와 예외 처리

### 커스텀 예외 (Custom Exception)

```python
num = int(input('아무 숫자나 입력해 보세요!!!'))

try :
    print(10 / num)
except :
    print('오잉~!!!')
```

```python
# 나만의 예외 클래스를 만들 수 있음!
class NotMyTeacher(Exception):
    pass
```

```python

teacher_name = input('당신의 선생님 이름을 입력하세요! : ')

if teacher_name == '다람쥐' :
    print('훌륭합니다!')
else :
    # raise를 이용해 예외 발생시킴(인자로 전달한 메시지가 에러 메시지가 됨.)
    raise NotMyTeacher('선생님 이름도 모르다니 이런..! ㅜㅠ')

```

### 예외처리 (try-except)

```python
def check_teacher(teacher_name) :
    if teacher_name == '다람쥐' :
        print('훌륭합니다!')
    else :
        raise NotMyTeacher('선생님 이름도 모르세요?..')

try :
    check_teacher('다람쥐')
    check_teacher('호랑이')
    print('선생님 이름 확인 환료!!')
except :
    print('예외가 발생했습니다! 악@@!!!!!')
```

```python
try :
    test_list = [2026, 1, 7]
    print(test_list[21])
    print(2026/10)
    raise NotMyTeacher()
# 특정 예외를 지정해서 처리할 수 있음
# as e를 붙여서 e에 예외 메시지를 담아 쓸 수 잇음.
# exception의 타입에는 계층 구조와 순서가 있음. 유의!
except IndexError as e :
    print('XXX 인덱싱 범위를 초과했습니다 XXX')
    print(e)
except ZeroDivisionError :
    print('XXX 0으로는 나눌 수 없습니다 XXX')
except :
    print('아무튼 예외가 발생했습니다! 예외 처리를 했어요!')
```

```python
 # 언제나 실행됨. 예외 발생시, 오류 발생시 try 수행 코드 즉시 정지
try :
    test_list = [2026, 1, 7]
    print(test_list[1])
except :
    print('아무튼 예외가 발생했습니다! 예외 처리를 했어요!')
else :
    print("else는 예외가 발생하지 않았을때 출력됨")
finally :
    print('finally는 예외 발생여부에 상관없이 동작이 끝나면 마무리로 실행됨.')
```
