```python
# 별찍기를 해보자 1
# *****
# ****
# ...

a = 5
for n in range(a) :
    print("*" * (a-n))
```

```python
#      *
#     **
#    ***
#   ****
#  *****
a = 6
for i in range(1, a):
    print(" " * (a - i), end="")
    print('*' * i)
```

```python
# 별찍기를 해보자 2
#    *
#   ***
#  *****
#   ***
#    *

a = 7
for i in range(a//2):       # 몫만큼 돌기
    print(' ' * (a//2 - i), end = '')
    print('*' * (2*i+1))

for i in range(a//2-1):       # 몫-1만큼 돌기
    print(' ' * (i + 2), end = '')
    print('*' * ((a//2*2)-3-2*i))
```

```python
# 개인 실습 student가 아닌 사람은 간식 주면 안되는 메서드 만들기!
user_list = [{"type":"student", "name":"김수진"}
             , {"type":"student", "name":"김김김"}
             ,{"type":"student", "name":"수수수"}
             , {"type":"teacher", "name":"남윤진"}
             , {"type":"manager", "name":"변수현"}]

# 사용자 타입 딕셔너리
type_dict = {"student":"학생", "teacher":"강사", "manager":"매니저"}

def snack_machine(list) :
    for n in list :
        if n['type'] == "student" :
            print(f'{n['name']}님에게 간식 전달 완료.')
        else :
            print(f'{n['name']}님은 학생이 아니네요.', end=' ')
            print(f'{n['name']}님은 {type_dict[n['type']]}입니다!')

snack_machine(user_list)
```
